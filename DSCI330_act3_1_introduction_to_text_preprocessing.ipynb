## Basic Text Cleaning

Common tasks:
1. Lowercasing,
2. Punctuation removal,
3. Tokenization: unfolding into sentences, words, or characters.

# Text Cleaning Basics - **Lowercasing**

**Why?** Making all words the same case removes case as a concern.

**How?** Use the `.lower` method of all strings.

**Careful!** Be sure to perform any processing, like proper name removal, before lowercasing!

```python
(persuasion_lower := persuasion_raw.lower())
```

```python
(persuasion_lower :=
 persuasion_raw
 >> pipeable(lambda s: s.lower())
)
```

# Text Cleaning Basics - **Punctuation Removal**

**Why?** Punctuation can make comparisons hard, e.g., `"Silas"` VS `"Silas."` VS. `"Silas!"`

**How?**
1. Make a punctuation translation `dict`
2. Use the string `translate` method.

**Careful!** Be sure to perform any processing, like sentence tokenization, before removing punctuation!

```python
punctuation = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
```

```python
from string import punctuation

(punc_map := str.maketrans('', '', punctuation))
```

```python
(persuasion_no_punc := persuasion_raw.translate(punc_map))
```

```python
# Functional Approach
(persuasion_no_punc :=
 persuasion_raw
 >> pipeable(lambda s: s.translate(punc_map))
)
```

# Text Cleaning Basics - **Tokenization** - (unfolding)

The act of *tokenizing* a text consists of breaking it into one of the levels of abstraction, typically

1. Sentences [e.g, Understand meaning],
2. Words [e.g, Measure sentiment], or
3. Characters [e.g, Text to speech].

#### Example 1 - **Tokenize into words**

```python
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize

(persuasion_words := word_tokenize(persuasion_raw))[:10]
```

#### Example 2 - **Tokenize into sentences**

```python
from nltk.tokenize import sent_tokenize

(persuasion_sents := sent_tokenize(persuasion_raw))[:10]
```

#### Example 3 - **Tokenize into characters**

```python
(persuasion_chrs := [ch for ch in persuasion_raw])[:10]
```

### Why we like piping $\longrightarrow$ composability

```python
(persuasion_clean_words :=
 persuasion_raw
 >> pipeable(lambda s: s.lower())
 >> pipeable(lambda s: s.translate(punc_map))
 >> pipeable(word_tokenize)
)[:10]
```

# Text Cleaning: **Stop word removal, Stemming, and Lemmatization**

When working with word tokens, we commonly perform the following preprocessing

1. **Stop word removal.** Remove common words that add little meaning.
2. **Stemming.** Reduce words to their base by removing suffixes/prefixes/affixes.
3. **Lemmatization.** Reduce words to their base or dictionary form.

### Why perform **stop word removal**?

1. Dimensional reduction: Lowering the size of the "vocabulary" reduces the number of words to compare.  Without stop word removal, these common words will over-shadow more meaningful terms.
2. **Processing Speed.** Less words to process $\longrightarrow$ faster processing.
3. **Better preformance/accuracy.** More meaningful words $\longrightarrow$ better machine learning performance.

**Sample stop words:** the, is, in, and


```python
from nltk.corpus import stopwords
nltk.download('stopwords') # Downloads the stopwords dataset if not already downloaded

eng_stop_words = set(stopwords.words('english')) # Gets the set of English stop words
```

```python
# keep things not in the stop words
(meaningful_words := [w for w in persuasion_words if w not in eng_stop_words])[:10]
```

```python
(meaningful_words :=
 persuasion_words
 >> filter(lambda w: w not in eng_stop_words)
)[:10]
```

### Text Cleaning - **Stemming**

```python
from nltk.stem import PorterStemmer # upper case P because this is "class", "function" is lowercase

stemmer = PorterStemmer()
```

```python
(stem_words := [stemmer.stem(w) for w in  persuasion_clean_words])[:10]
```

```python
# Investigate the difference after stemming
(stem_words := [(w, stemmer.stem(w)) for w in  persuasion_clean_words])[:30]
```

```python
(stem_words :=
 persuasion_clean_words
 >> map(lambda w: stemmer.stem(w))
)[:10]
```

### Text Cleaning - **Lemmatizing**

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()
```

```python
(lemma_words := [lemmatizer.lemmatize(w) for w in  persuasion_clean_words])[:10]
```

```python
# Investigate the difference after lemmatizing
(lemma_words := [(w, lemmatizer.lemmatize(w)) for w in  persuasion_clean_words])[:30]
```

```python
(lemma_words :=
 persuasion_clean_words
 >> map(lambda w: lemmatizer.lemmatize(w))
)[:10]
```

# **Excercise 3.1**

Performe the following on Sense and Sensibility by Jane Austen

1. Download and load the text
2. Remove the title,
3. Preclean (make lower, remove punc, split into words)
4. Word cleaning (remove stop words, stem)

```python
import nltk
nltk.corpus.gutenberg.fileids()
```

```python
# 1. Download and load the text
(sense_raw := gutenberg.raw('austen-sense.txt'))[:2000]
```

```python
print(sense_raw[:2000])
```

```python
# 2. Removing the title
(sense_raw := gutenberg.raw('austen-sense.txt')[45:])[:2000]
```

```python
# 3. Preclean
# make lowercase
(sense_lower := sense_raw.lower())
```

```python
from string import punctuation

(punc_map := str.maketrans('', '', punctuation))
```

```python
# 3. Preclean
# remove punctuations
(sense_no_punc := sense_lower.translate(punc_map))
```

```python
# 3. Preclean
# split into words
from nltk.tokenize import word_tokenize

(sense_words := word_tokenize(sense_no_punc))[:20]
```

```python
# 4. Word cleaning
# remove stop words

from nltk.corpus import stopwords
nltk.download('stopwords') # Downloads the stopwords dataset if not already downloaded

eng_stop_words = set(stopwords.words('english')) # Gets the set of English stop words
print(eng_stop_words)
```

```python
# stop words such as 'the', 'of', 'in' have been removed from sense_words
(meaningful_words := [w for w in sense_words if w not in eng_stop_words])[:10]
```

```python
# 4. Word cleaning
# stemming

from nltk.stem import PorterStemmer # upper case P because this is "class", "function" is lowercase

stemmer = PorterStemmer()
```

```python
(stem_words := [stemmer.stem(w) for w in  meaningful_words])[:10]
```

```python
# Investigate the difference after stemming
# Words were normalized: family -> famili
(stem_words := [(w, stemmer.stem(w)) for w in  meaningful_words])[:30]
```
